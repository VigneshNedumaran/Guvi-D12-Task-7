# KNN

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from scipy.spatial import distance
```

```python
column_name = ['ID', 'RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type']
```

```python
train = pd.read_csv('trainKNN.txt',names= column_name)
```

```python
train.head()
```

```python
test = pd.read_csv('testKNN.txt',names= column_name)
```

```python
test.head()
```

```python
train = train.drop('ID', axis=1)
test = test.drop('ID', axis=1)
```

```python
train.describe()
```

```python
test.describe()
```

```python
def standardize (df):
    for col in df.columns:
        if col != "Type": 
            df[col] = (df[col] - df[col].mean())/df[col].std()
    return df
```

```python
train = standardize(train)
test = standardize(test)
train.head()
```

```python
euclid_model = KNeighborsClassifier(n_neighbors=8, metric=distance.sqeuclidean) 
manhattan_model = KNeighborsClassifier(n_neighbors=8, metric=distance.cityblock)
```

```python
x_train = train.drop(["Type"], axis=1)
y_train = train["Type"]
euclid_model.fit(x_train,y_train)
```

```python
x_test = test.drop("Type", axis=1) 
y_test = test["Type"]
manhattan_model.fit(x_train, y_train)
```

```python
manhattan_predictions = manhattan_model.predict(x_test)
euclid_predictions = euclid_model.predict(x_test) 
df = pd.DataFrame({'actual': y_test, 'manhattan': manhattan_predictions, 'euclid': euclid_predictions})
df.head()
```
